```python
import math
import random
import matplotlib.pyplot as plt

def simulation(n):
    # setting the figure size and aspect ratio
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111)
    ax.set_aspect(aspect = 'equal')
    
    # plot n random points
    points_in_circle = 0
    for i in range(0, n):
        x = random.random()
        y = random.random()
        
        distance_from_origin = math.sqrt(x**2 + y**2) # distance formula
        
        if distance_from_origin < 1:
            plt.plot(x, y, 'bo', markersize=1) # blue marker if less than 1
            points_in_circle += 1
        else:
            plt.plot(x, y, 'ro', markersize=1) # red else if greater than 1
    
    # estimate π and error rate
    pi = 4 * (points_in_circle / n)
    print("n = " + str(n), "pi = " + str(pi), "error = " + str(abs(math.pi - pi)) + "%")
    
    plt.title("Estimating π using simulation, n = " + str(n))
    plt.show()
```

```python
def main():
    n_values = [1000, 10000, 100000, 1000000]

    for n in n_values:
        simulation(n)
        
    #simulation(1000)
```

```python
main()
```
